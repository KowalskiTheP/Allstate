```python
#Correlation matix
#Es werden die Daten mit fortlaufender Kategorieumformung verwendet. 
#(Unter umständen kann man das aber auch ändern und letzlich nur Teile der Kategorien verwenden)
correlation_data = "../Data/conv_training.csv"
correlation_dataset = pd.read_csv(correlation_data,delimiter=",", header=0, index_col=0)
#print correlation_dataset.iloc[0,:]
dataset_npArray = correlation_dataset.as_matrix(columns=None)

hightCorr_array = []
lowCorr_array = []

covariance_matrix = np.cov(dataset_npArray.T)
correlation_matrix = np.corrcoef(dataset_npArray.T)
corr_array = correlation_matrix[-1]
for i in range(len(corr_array)):
  if abs(corr_array[i]) < 0.1:
    lowCorr_array.append(i)
    #print i,correlation_dataset.iloc[0,i], corr_array[i]
print "\n"
for i in range(len(corr_array)):
  if abs(corr_array[i]) > 0.1:
    hightCorr_array.append(i)
    #print i,correlation_dataset.iloc[0,i] , corr_array[i]

#np.savetxt('correlationMatrix.txt',np.corrcoef(dataset_npArray.T), delimiter=',', newline='\n')
print "Hohe Korrelation", hightCorr_array
print "Geringe Korrelation", lowCorr_array
```

```python
#Feature importances
import matplotlib.pylab as plt
%matplotlib inline
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 12, 4

xgtrain = xgb.DMatrix(final_train, label=y_train)

impTest = XGBRegressor(max_depth=3, 
                         learning_rate=0.05, 
                         n_estimators=3000, 
                         silent=True, 
                         objective='reg:linear', 
                         nthread=-1, 
                         gamma=0, 
                         min_child_weight=1, 
                         max_delta_step=0, 
                         subsample=1.0, 
                         colsample_bytree=1, 
                         colsample_bylevel=1, 
                         reg_alpha=0, 
                         reg_lambda=1, 
                         scale_pos_weight=1, 
                         base_score=0.5, 
                         seed=42, 
                         missing=None)

xgb_param = impTest.get_xgb_params()
cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=impTest.get_params()['n_estimators'], nfold=5,
            metrics='mae', early_stopping_rounds=100)
print cvresult

impTest.fit(final_train,y_train)

importances = pd.Series(impTest.booster().get_fscore()).sort_values(ascending=False)
importances_index = importances.index
#print importances_index
final_train_columns = final_train.columns
#print len(final_train_columns)
#print final_train_columns

#Diese Schleife erweitert importances um jene Feature mit importance 0 (werden von xgBoost nicht mit aufgenommen).
for i in range(len(final_train_columns)):
    if final_train_columns[i] not in importances_index:
        s3 = pd.Series([0], index=[final_train_columns[i]])
        importances = importances.append(s3, verify_integrity=True)

feat_imp = pd.Series(impTest.booster().get_fscore()).sort_values(ascending=False)
feat_imp.plot(kind='bar', title='Feature Importances')
plt.ylabel('Feature Importance Score')
```

```python
unImpXGB = []
for i in range(len(importances)):
    if importances[i]<=00.0:
        unImpXGB.append(str(importances.index[i]))
print len(final_train.columns)
print len(unImpXGB), unImpXGB
```

```python
#Abgleich zischen Korrelationsmatrix und feature imortance. 
#Wenn die feature importance unter einem bestimmten Wert liegt und die Oberkategorie im array der geringen Korrelation vorhanden ist, 
#dann wird diese Kategorie im toDrop_StrArray angezeigt.

lowCorr_StrArray = []
toDrop_StrArray = []
print len(importances)
for i in range(len(lowCorr_array)):
    lowCorr_StrArray.append("cat"+str(lowCorr_array[i]))
    
for i in range(len(importances)):
    fImpStr = str(importances.index[i]).split("_")[0]
    if fImpStr in lowCorr_StrArray and importances[i] < 15:
        if fImpStr not in toDrop_StrArray:
            toDrop_StrArray.append(fImpStr)
            
print toDrop_StrArray
```

```python
cv_params = {'max_depth': [1,3,5], 'min_child_weight': [1,3,5]}
ind_params = {'learning_rate': 0.1, 'n_estimators': 1000, 'seed':42, 'subsample': 0.8, 'colsample_bytree': 0.8, 
             'objective': 'reg:linear'}
optimized_GBM = GridSearchCV(xgb.XGBRegressor(**ind_params), 
                            cv_params, 
                            scoring = 'neg_mean_absolute_error', 
                            cv = 2, 
                            n_jobs = -1)

print "Check 1"
# Optimize for accuracy since that is the metric used in the Adult Data Set notation
#optimized_GBM.fit(final_train, y_train)
print "Check 2"
gsearch1 = GridSearchCV(cv=5, 
             error_score='raise', 
             estimator= XGBRegressor(max_depth=3, 
                                     learning_rate=0.1, 
                                     n_estimators=1000, 
                                     silent=True, 
                                     objective='reg:linear', 
                                     nthread=4, 
                                     gamma=0, 
                                     min_child_weight=7, 
                                     max_delta_step=0, 
                                     subsample=1, 
                                     colsample_bytree=1, 
                                     colsample_bylevel=1, 
                                     reg_alpha=0, 
                                     reg_lambda=1, 
                                     scale_pos_weight=1, 
                                     base_score=0.5, 
                                     seed=42, 
                                     missing=None),
             fit_params={}, 
             iid=True, 
             n_jobs=1,
             param_grid={'min_child_weight': [1,2,3,4,5], 'max_depth': [2,3,4,5]},
             #pre_dispatch='2*n_jobs', 
             #pre_dispatch=2, 
             refit=True, 
             scoring='neg_mean_absolute_error', 
             verbose=3)

gsearch1.fit(final_train,y_train)


```
